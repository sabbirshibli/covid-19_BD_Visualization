```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams
sb.set_style('whitegrid')
```

```python
%matplotlib inline
rcParams['figure.figsize']=5,4
```

```python
address = 'COVID-19_BD_Dataset-8_april.csv'
```

```python
bdc = pd.read_csv(address)
bdc.head()
```

```python
bdc
```

```python
dtest = bdc['Daily New Tests']
dconfirmed = bdc['Daily new confirmed cases']
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(dtest)
ax2.plot(dconfirmed)
```

```python
color_theme = ['red']
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
dconfirmed.plot(color=color_theme)
ax.set_xticks(range(32))
ax.set_xticklabels(bdc.Date, rotation=70, fontsize='medium')
ax.set_title('Daily Confirmed Cases of COVID-19 (Bangladesh)')
ax.set_xlabel('Dates')
ax.set_ylabel('Confirmed Cases')
ax.legend(loc='best')
plt.savefig('BD_Confirmed_Cases.png')
```
